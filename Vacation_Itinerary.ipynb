```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import gmaps

import requests
from config import g_key
```

```python
#Import hotel df
gmaps.configure(api_key=g_key)

hotel_df = pd.read_csv('data/WeatherPy_vacation.csv')

hotel_df.head(10)
```

```python
#Select cities for travel

cities = ['Naron', 'Carballo', 'Muros', 'Agucadoura']
```

```python
#Create dataFrame for cities to be traveled to

naron_df = hotel_df.loc[(hotel_df['City'] == 'Naron')]
carballo_df = hotel_df.loc[(hotel_df['City'] == 'Carballo')]
muros_df = hotel_df.loc[(hotel_df['City'] == 'Muros')]
agucadoura_df = hotel_df.loc[(hotel_df['City'] == 'Agucadoura')]

naron_df.head()
```

```python
# get latitude and longitude pairs

naron_lat_lng = []
naron_df.values[0][5]
naron_lat_lng.append(naron_df.values[0][5])
naron_lat_lng.append(naron_df.values[0][6])
naron_lat_lng = tuple(naron_lat_lng)

carballo_lat_lng = []
carballo_lat_lng.append(carballo_df.values[0][5])
carballo_lat_lng.append(carballo_df.values[0][6])
carballo_lat_lng = tuple(carballo_lat_lng)

muros_lat_lng = []
muros_lat_lng.append(muros_df.values[0][5])
muros_lat_lng.append(muros_df.values[0][6])
muros_lat_lng = tuple(muros_lat_lng)

agucadoura_lat_lng = []
agucadoura_lat_lng.append(agucadoura_df.values[0][5])
agucadoura_lat_lng.append(agucadoura_df.values[0][6])
agucadoura_lat_lng = tuple(agucadoura_lat_lng)
```

```python
travel_df = pd.concat([naron_df, carballo_df, muros_df, agucadoura_df])

travel_df.head()
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""

# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in travel_df.iterrows()]
```

```python
# Add a map for the trave route.

fig = gmaps.figure()

naron2agucadoura = gmaps.directions_layer(naron_lat_lng, agucadoura_lat_lng, waypoints=[carballo_lat_lng, muros_lat_lng], travel_mode='DRIVING')
fig.add_layer(naron2agucadoura)

# Call the figure to plot the data.
fig

```

```python
fig = gmaps.figure()

locations = travel_df[["Lat", "Lng"]]

marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)
                      
fig
```
